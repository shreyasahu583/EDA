```python
# Task 2
#1
sns.histplot(df["age"], kde=True).set_title("Distribution of Age")
plt.show()

sns.histplot(df["watch_hours"], kde=True).set_title("Distribution of Watch Hours")
plt.show()

sns.histplot(df["monthly_fee"], kde=False).set_title("Distribution of Monthly Fee")
plt.show()

sns.countplot(x="churned", data=df).set_title("Churn Count")
plt.xticks([0, 1], ["No Churn", "Churn"])
plt.show()
```

```python
#2
for col in cat_cols:
    sns.countplot(x=col, data=df, order=df[col].value_counts().index)
    plt.title(f"Count Plot for {col}")
    plt.xticks(rotation=45)
    plt.show()
```

```python
# Task 3
#1
for col in ["subscription_type", "region", "device"]:
    grouped = df.groupby(col)[["watch_hours", "monthly_fee"]].mean().reset_index()
    
    sns.barplot(x=col, y="watch_hours", data=grouped)
    plt.title(f"Average Watch Hours by {col}")
    plt.xticks(rotation=45)
    plt.show()
    
    sns.barplot(x=col, y="monthly_fee", data=grouped)
    plt.title(f"Average Monthly Fee by {col}")
    plt.xticks(rotation=45)
    plt.show()
```

```python
#2
genre_watch = df.groupby("favorite_genre")["avg_watch_time_per_day"].mean().reset_index()
sns.barplot(x="favorite_genre", y="avg_watch_time_per_day", data=genre_watch,
            order=genre_watch.sort_values("avg_watch_time_per_day", ascending=False)["favorite_genre"])
plt.title("Avg Watch Time Per Day by Genre")
plt.xticks(rotation=45)
plt.show()
```

```python
#3
for col in ["gender", "region", "subscription_type", "payment_method"]:
    churn_rate = df.groupby(col)["churned"].mean().reset_index()
    sns.barplot(x=col, y="churned", data=churn_rate)
    plt.title(f"Churn Rate by {col}")
    plt.ylabel("Churn Rate")
    plt.xticks(rotation=45)
    plt.show()
```

```python
# Task 5
#1
insights = []
# Insight 1: Watch hours & churn
watch_churn = df.groupby("churned")["watch_hours"].mean()
if watch_churn[1] < watch_churn[0]:
    insights.append("Customers with lower watch hours are more likely to churn.")

# Insight 2: Subscription type churn
sub_churn = df.groupby("subscription_type")["churned"].mean()
insights.append(f"'{sub_churn.idxmin()}' users churn the least, '{sub_churn.idxmax()}' churn the most.")

# Insight 3: Region with highest average watch hours
region_watch = df.groupby("region")["watch_hours"].mean()
insights.append(f"Region with highest average watch hours: {region_watch.idxmax()}")

# Insight 4: Device with highest average watch hours
device_watch = df.groupby("device")["watch_hours"].mean()
insights.append(f"Device with highest average watch hours: {device_watch.idxmax()}")

# Insight 5: Genre with highest avg watch time per day
genre_watch = df.groupby("favorite_genre")["avg_watch_time_per_day"].mean()
insights.append(f"Genre with highest avg watch time per day: {genre_watch.idxmax()}")

# Insight 6: Correlation with churn
corr_matrix = df[["age", "watch_hours", "last_login_days", "monthly_fee",
                  "number_of_profiles", "avg_watch_time_per_day", "churned"]].corr()
if abs(corr_matrix["churned"]["watch_hours"]) > 0.2:
    insights.append("Watch hours show noticeable correlation with churn.")
if abs(corr_matrix["churned"]["last_login_days"]) > 0.2:
    insights.append("Last login days show correlation with churn.")

# Print insights
for i, insight in enumerate(insights, 1):
    print(f"Insight {i}: {insight}")
```
